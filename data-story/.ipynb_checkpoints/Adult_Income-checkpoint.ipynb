## Pre-processing Data

We will drop the rows that have '?' because we do not have supplemental information to impute this missing value.

```python
#Drop rows with missing values
adult = adult.replace('?', np.nan)
adult = adult.dropna()
adult.shape
```

## Relationship Between Income and Age, Year of Education and Hours Per Week

```python
#Plot relationship between age, years of education and hours per week on income level. 
plt.figure(figsize = (13,11))
plt.subplot(2, 2, 1)
sns.boxplot(x = 'income', y = 'age', data = adult)
plt.title('Plot of age versus income')

plt.subplot(2, 2, 2)
sns.boxplot(x = 'income', y = 'educational-num', data = adult)
plt.title('Plot of years of education versus income')

plt.subplot(2, 2, 3)
sns.boxplot(x = 'income', y = 'hours-per-week', data = adult)
plt.title('Plot of hours-per-week versus income')


plt.tight_layout()
plt.show()
```

We make the following observations on the the relationship between age, number of years of education and hours-per-week worked on income levels.

- Older people tend to have higher incomes than younger people.
- Those with higher number of years of education make more money than those with less number of years of education. In fact, it seems that the middle 50% of individuals making over \\$50K have a higher of years of education make more money than the midspread of making less than \\$50K.
- We observe a similar result for those who work longer hours. The midspread of those making over \\$50K tend to work longer than the midspread of those earning less than \\$50K.


## Relationship Between Income and Capital Gain Or Losses

```python
#Plot relationship between capital gain / loss and income level
plt.figure(figsize = (13,8))
plt.subplot(1, 2, 1)
sns.boxplot(x = 'income', y = 'capital-loss', data = adult)
plt.title('Plot of capital-gain versus income')

plt.subplot(1, 2, 2)
sns.boxplot(x = 'income', y = 'capital-gain', data = adult)
plt.title('Plot of capital-loss versus income')

plt.tight_layout()
plt.show();
```

We first of all note that most people don't realize capital-gains or losses, and those that do are the exception, inspite of income-level. This may imply that most people don't invest. We also note that individuals who earn over and under \\$50K are equally likely to realize capital gains and losses, although it appears that those with incomes over \\$50K actualize higher magnitude of gains than those with incomes less than \\$50K. Furthermore, we observe that both income classes generally actualize higher magnitudes of capital gains than losses.

## Does the type of work you do determine the hours of work?

```python
plt.figure(figsize = (13,8))
hours_of_work = adult.groupby(['occupation'])[['hours-per-week']].mean().sort_values(by = 'hours-per-week').reset_index()
sns.barplot(x = 'hours-per-week', y = 'occupation', data = hours_of_work)
plt.title('Occupation versus hours-per-week')
plt.axvline(x =40);
```

```python
plt.figure(figsize = (13,10))
sns.boxplot(data = adult, x = 'hours-per-week', y = 'occupation', hue = 'income')
plt.title('Relationship between hours worked per week, occupation and income-level');
```

We observe than on average executives and farmers/fishermen work more than other occupations. However, there is a lot of variation associated with working long hours and making over \\$50K, suggesting that longer hours may not necessarily translate to higher incomes. That said, the 40-hour mark seems to be a relevant threshold below which individuals are more likely to make incomes higher than \\$50K, the only exception being private house servants. We also note that for most occupations, over 75% of individuals work over 40 hours a week. 

## How does Education Affect Investing Behavior

```python
plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
sns.boxplot(x = 'educational-num', y = 'capital-gain', data = adult)
plt.title('Plot of capital gain versus years of education')

plt.subplot(1,2,2)
sns.boxplot(x = 'educational-num', y = 'capital-loss', data = adult)
plt.title('Plot of capital loss versus years of education')


plt.tight_layout()
plt.show()
```

As noted, most people do not actualize capital-gains or losses and we observe this trend inspite of their number of years of education. 

```python
capital_gain_education = adult.groupby(['educational-num'])[['capital-gain']].mean();
capital_loss_education = adult.groupby(['educational-num'])[['capital-loss']].mean();

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5))
capital_gain_education.plot(ax = ax1, title = 'Mean capital gain versus education-level', ylabel = 'capital-gain')
ax1.axvline(x = 12, ls = '--', color = 'r')
capital_loss_education.plot(ax = ax2, title = 'Mean capital loss versus education-level', ylabel = 'capital-loss')
ax2.axvline(x = 12, ls = '--', color = 'r')
plt.tight_layout()
plt.show()
```

We observe that on average individuals that chose to invest are equally likely to realize high capital-gains and capital-losses if they have finished high school. This may suggest that individuals with less than a high school education take less investment risks than those with more educaiton. 

```python
capital_gain_education = adult.groupby(['educational-num'])[['capital-gain']].sum();
capital_loss_education = adult.groupby(['educational-num'])[['capital-loss']].sum();

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5))
capital_gain_education.plot(ax = ax1, title = 'Total capital gain versus education-level', ylabel = 'capital-gain')
ax1.axvline(x = 12, ls = '--', color = 'r')
capital_loss_education.plot(ax = ax2, title = 'Total capital loss versus education-level', ylabel = 'capital-loss')
ax2.axvline(x = 12, ls = '--', color = 'r')
ax1.axvline(x = 8, ls = '--', color = 'r')
ax2.axvline(x = 8, ls = '--', color = 'r')
plt.tight_layout()
plt.show()
```

We note that there are two education thresholds where investing activity picks up - past 8 years and past high school education. It also seems that those with more education past high-school tend to actualize more capital-gains and reduce their capital-losses when compared to those with education levels between 8 and 12 years.
