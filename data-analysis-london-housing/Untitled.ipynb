```python
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
prices = pd.read_excel(url_LondonHousePrices, sheet_name = 'Average price')
```

```python
prices.head()
```

```python
price = prices.T.reset_index()
```

```python
price.head()
```

```python
price.columns = price.iloc[0]
```

```python
price.drop(index = [0], inplace = True)
```

```python
price.head()
```

```python
price = price.rename({'Unnamed: 0': 'borough'}, axis = 1)
price.head()
```

```python
price.columns = price.columns.fillna('id')
price.head()
```

```python
properties = pd.melt(price, id_vars = ['borough', 'id'], var_name = 'date', value_name = 'avg_price')
```

```python
properties
```

```python
properties[properties.isnull().any(axis = 1)]
```

```python
properties = properties.dropna(axis = 0)
```

```python
properties.shape
```

```python
properties.head(50)
```

```python
real_boroughs = ['Barking & Dagenham', 'Barnet', 'Bexley',
       'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield',
       'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Haringey',
       'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington',
       'Kensington & Chelsea', 'Kingston upon Thames', 'Lambeth',
       'Lewisham', 'Merton', 'Newham', 'Redbridge',
       'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets',
       'Waltham Forest', 'Wandsworth', 'Westminster']
```

```python
len(real_boroughs)
```

```python
properties.borough.nunique()
```

```python
properties = properties[(properties.borough.isin(real_boroughs))]
properties.head()
```

```python
properties.shape
```

```python
properties[properties.borough == 'Barnet']
```

```python
properties['year'] = properties.date.dt.year
```

```python
properties.avg_price = round(properties.avg_price.astype('float'), 2)
```

```python
properties_group  = properties.groupby(['borough', 'year'])['avg_price'].agg(mean_price = ('mean')).apply(lambda x: round(x, 2))
```

```python
properties_group.head(10)
```

```python
properties_df = properties_group.reset_index()
properties_df.head(30)
```

```python
properties_group.index.names
```

```python
a = properties_df[properties_df['borough'] == 'Barnet'].year.max()
a
```

```python
properties_df.loc[(properties_df['borough'] == 'Barnet') & (properties_df.year == a), 'mean_price'].iloc[0]
```

```python
perc_change_dict = defaultdict(float)
for bor in real_boroughs:
    min_year = properties_df[properties_df['borough'] == bor].year.min()
    max_year = properties_df[properties_df['borough'] == bor].year.max()
    price_min_year = properties_df.loc[(properties_df['borough'] == bor) & 
                                       (properties_df.year == min_year), 'mean_price'].iloc[0]
    price_max_year = properties_df.loc[(properties_df['borough'] == bor) & 
                                       (properties_df.year == max_year), 'mean_price'].iloc[0]
    perc_change = ((price_max_year - price_min_year) / price_min_year) * 100
    perc_change_dict[bor] = perc_change    
```

```python
perc_change_dict
```

```python
df = pd.DataFrame.from_dict(perc_change_dict, orient = 'index', columns = ['26_yr_change'])
df
```

```python
df.loc[df['26_yr_change'] == df['26_yr_change'].max()]
```

```python
change20_dict  = defaultdict(float)
for bor in real_boroughs:
    price_2000 = properties_df.loc[(properties_df['borough'] == bor) & 
                                       (properties_df.year == 2000), 'mean_price']
    price_2020 = properties_df.loc[(properties_df['borough'] == bor) & 
                                       (properties_df.year == 2020), 'mean_price']
    change20 = ((price_2020 - price_2000) / price_2000) * 100
    change20_dict[bor] = change20    
```

```python
properties_df['mean_price'][properties_df.year == 2000]
```

```python
df20 = pd.DataFrame.from_dict(change20_dict, orient = 'index', columns = ['20_yr_change'])
df20
```

```python
df20.loc[df20['20_yr_change'] == df20['20_yr_change'].max()]
```
